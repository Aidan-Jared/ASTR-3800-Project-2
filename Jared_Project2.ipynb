```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits as pyfits
import scipy.stats as sc
%matplotlib inline
import Jared_function as ai
```

# A

###### i

```python
ki = pyfits.open('frame-g-000094-1-0131.fits') #takes the fits file data into an array
z = ki[0].data
#
#
z[z < 0] = np.min(np.abs(z))# removes noise from the data
z = np.log10(z) #takes the log
plt.imshow(z, cmap='Greys_r', interpolation='bicubic', extent=[354.368,354.577, -1.00631, -.852765], origin = 'lower')
plt.colorbar()
ai.figName('Fig 1: Fits file image','RA', 'DEC')
plt.show()
ai.pause()
```

###### ii

```python
x, y, z = ai.csvArray('frame-g-000094-1-0131.csv') #takes the csv data into an array
x1 = x * 1.025e-4 + 354.368 #changes from pixels to RA and DEC
y1 = y * 1.025e-4 - 1.00631
z = z/1e3
z = 10**(z / -2.5) #calculates the magnitude of the stars
plt.scatter(x1,y1, s = z*5)
plt.xlim([min(x1),max(x1)])
plt.ylim([min(y1),max(y1)])
ai.figName('Fig 2: CSV Star Field','RA', 'DEC')
plt.show()
ai.pause()
```
